```python
# Import required packages for feature extraction
import radiomics
from radiomics import featureextractor
import pandas as pd
```

```python
# Function to iterate folders for feature extraction
def extract_features (root_dir, extractor):
  results=[]

  for dirName, _, fileList in os.walk(root_dir):
    for fname in fileList:
      if 't1_LPS_rSRI' in fname:
        folder_name=os.path.basename(dirName)
        image_path=os.path.join(dirName, fname)
        label_fname=f"{folder_name}_GlistrBoost_out.nii.gz"
        label_path=os.path.join(dirName, label_fname)

        if os.path.exists(label_path):
          result=extractor.execute(image_path, label_path)
          result["id"]=folder_name
          results.append(result)

  df=pd.DataFrame(results)
  df.set_index('id', inplace=True)

  return df
```

```python
rootDir='/content/drive/MyDrive/HIDS_7009/7009_Project/Rembrandt_Final_Data'
extractor = featureextractor.RadiomicsFeatureExtractor()
```

```python
df=extract_features(rootDir, extractor)
```
